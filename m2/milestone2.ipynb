```python
# importing the required module 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
  
# read csv file
walking = pd.read_csv('../lab8-dataset/WALKING.csv')
walking
```

```python
walking['smoothaccelerationx'] = 0.5 * walking['accel_x'] + (1 - 0.5) * (walking['accel_x'] - 1)

walking
```

```python
plt.plot(walking['timestamp'], walking['accel_x'], label = 'Acceleration X')
plt.plot(walking['timestamp'], walking['smoothaccelerationx'], label = 'Smooth Acceleration X')
plt.xlabel('Timestamp (s)')
plt.ylabel('Acceleration (m/s²)')
plt.title('Acceleration and Time Data')
plt.legend()
plt.show()
```
