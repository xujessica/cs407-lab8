```python
# importing the required module 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
  
# read csv file
acceleration = pd.read_csv('../lab8-dataset/ACCELERATION.csv')
acceleration
```

```python
plt.plot(acceleration['timestamp'], acceleration['acceleration'], label = 'Acceleration')
plt.plot(acceleration['timestamp'], acceleration['noisyacceleration'], label = 'Noisy Acceleration')
plt.xlabel('Timestamp (s)')
plt.ylabel('Acceleration and Noisy Acceleration (m/sÂ²)')
plt.title('Acceleration and Time Data')
plt.legend()
plt.show()
```

```python
from scipy import integrate
# integration of acceleration to get speed
acceleration['speed'] = integrate.cumtrapz(acceleration['acceleration'], acceleration['timestamp'], initial = 0)
acceleration['noisyspeed'] = integrate.cumtrapz(acceleration['noisyacceleration'], acceleration['timestamp'], initial = 0)
acceleration

plt.plot(acceleration['timestamp'], acceleration['speed'], label = 'Speed')
plt.plot(acceleration['timestamp'], acceleration['noisyspeed'], label = 'Noisy Speed')
plt.xlabel('Timestamp (s)')
plt.ylabel('Speed and Noisy Speed (m/s)')
plt.title('Speed and Time Data')
plt.legend()
plt.show()
```

```python
# integration of speed to get distance
acceleration['distance'] = integrate.cumtrapz(acceleration['speed'], acceleration['timestamp'], initial = 0)
acceleration['noisydistance'] = integrate.cumtrapz(acceleration['noisyspeed'], acceleration['timestamp'], initial = 0)
acceleration

plt.plot(acceleration['timestamp'], acceleration['distance'], label = 'Distance')
plt.plot(acceleration['timestamp'], acceleration['noisydistance'], label = 'Noisy Distance')
plt.xlabel('Timestamp (s)')
plt.ylabel('Distance and Noisy Distance (m)')
plt.title('Distance and Time Data')
plt.legend()
plt.show()
```
